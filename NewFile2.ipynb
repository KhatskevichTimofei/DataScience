```python
import pandas as pd
```

```python
money = pd.read_excel("usd_exchange_rate.xlsx")
```

```python
money
```

```python
past = 7 * 4 #на основе за последние 4 недели 
future = 7 #будем предсказывать тренд на неделю вперёд
```

```python
values = money.curs
```

```python
start = past
end = len(values) - future
```

```python
print(start, end)
```

```python
raw_df = []
for i in range(start, end):
    #Для каждого дня строчки(т.е для каждого дня)
    past_and_future_values = values[(i-past):(i+future)]#Выбирает данные за прошлое и за будущее
    raw_df.append(list(past_and_future_values))
```

```python
values[0:7]
```

```python
raw_df 
```

```python
past_columns = [f"past_{i}" for i in range(past)] #Название для калонок в прошлом
```

```python
future_columns = [f"future_{i}" for i in range(future)] #Название для колонок в будущем
```

```python
df = pd.DataFrame(raw_df, columns = (past_columns + future_columns))
```
